```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresultState-369.htm"
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text,'html')

print(soup)
```

```python
soup.find('table')
```

```python
BJP_Table = soup.find('table',class_ ='table table-striped table-bordered')
```

```python
BJP_Table
```

```python
Headings=soup.find_all('th')
```

```python
Headings
```

```python
table_titles = [title.text.strip() for title in Headings]
print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
columns_data=soup.find_all('tr')
```

```python
columns_data
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('BJP.csv', index=False)
```
